```python
meanX = df['X'].mean()
print(meanX)
stdX = df['X'].std()/20
interval=stats.norm.interval(0.95,meanX,stdX)  
print('X 95% confidence intervals for the mean{}'.format(interval))
```

```python
meanY = df['Y'].mean()
print(meanY)
stdY = df['Y'].std()/20
interval=stats.norm.interval(0.95,meanY,stdY)  
print('Y 95% confidence intervals for the mean{}'.format(interval))
```
